```python
import pandas as pd
```

```python
# Downloaded from https://www.smard.de/en/downloadcenter/download-market-data/?downloadAttributes=%7B%22selectedCategory%22:1,%22selectedSubCategory%22:1,%22selectedRegion%22:%22DE%22,%22selectedFileType%22:%22CSV%22,%22from%22:1661983200000,%22to%22:1696975199999%7D
df = pd.read_csv('data/Actual_generation_202209010000_202310102359_Hour.csv', sep=';')

# Create index from time. TODO: convert to datetime
df['Time'] = df.Date + ' ' + df.Start 
df = df.set_index(pd.to_datetime(df['Time']))
del df['Date']
del df['Start']
del df['End']
del df['Time']

# Convert all cols to floats. Remove commas separating thousands
for col in df.columns:
    df[col] = df[col].astype(str).str.replace(',', '').astype(float)

display(df.head())
df.shape
```
